```python
import pandas as pd
```

```python
df = pd.read_csv('data/data/00_tracks.csv') 
```

```python
df['laneletId'].unique()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

print("Rozpoczynam Krok 2a: Wizualizacja mapy pas√≥w (lanelets)...")

if 'df' in locals() and df is not None:
    # We≈∫ tylko u≈Çamek danych (np. 1%), ≈ºeby wykres wygenerowa≈Ç siƒô szybko
    sample_df = df.sample(frac=0.01, random_state=42)
    
    # U≈ºyjemy 'hue', aby pokolorowaƒá punkty na podstawie 'laneletId'
    # U≈ºyjemy 'legend=False', bo legenda by≈Çaby gigantyczna i nieczytelna
    plt.figure(figsize=(15, 10))
    sns.scatterplot(
        data=sample_df, 
        x='xCenter', 
        y='yCenter', 
        hue='laneletId', 
        palette='tab20',  # U≈ºyj palety kolor√≥w, aby rozr√≥≈ºniƒá pasy
        s=1,              # Ma≈Çe punkty
        legend=False      # Wy≈ÇƒÖcz legendƒô
    )
    
    plt.title('Wizualizacja segment√≥w drogi (Lanelets) na podstawie pozycji pojazd√≥w', fontsize=16)
    plt.xlabel('Pozycja X [m]', fontsize=12)
    plt.ylabel('Pozycja Y [m]', fontsize=12)
    plt.axis('equal') # Zachowaj proporcje osi, aby zobaczyƒá prawdziwy kszta≈Çt drogi
    plt.show()

else:
    print("B≈ÇƒÖd: 'tracks_df' nie istnieje. Uruchom najpierw Kom√≥rkƒô 2, aby wczytaƒá dane.")
```
